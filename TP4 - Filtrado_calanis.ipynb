$$ \hat{b} = \mathop{med}_{600} \{ \mathop{med}_{200} \{s\} \} $$

Donde $s$ representa la señal de ECG a partir de la que estimamos el movimiento de la línea de base $ \hat{b}$. El operador $\mathop{med}_{200}$ calcula muestra a muestra la mediana en los 200 ms previos a la muestra actual. Luego dicha estimación la substraemos para estimar la actividad eléctrica cardíaca $\hat{x}$

$$ \hat{x} = s - \hat{b} $$

```python

##########################################
# Acá podés generar los gráficos pedidos #
##########################################
import warnings
warnings.filterwarnings('ignore')

# Módulos importantantes
import scipy.signal as sig
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.io as sio


fig_sz_x = 14
fig_sz_y = 13
fig_dpi = 80 # dpi

fs=1000

fig_font_family = 'Ubuntu'
fig_font_size = 16
#Diseño de los filtros digitales

# para listar las variables que hay en el archivo
#io.whosmat('ecg.mat')
mat_struct = sio.loadmat('ECG_TP4')

ecg_one_lead = mat_struct['ecg_lead']
ecg_one_lead = ecg_one_lead.flatten(1)
cant_muestras = len(ecg_one_lead)
nyq_frec = fs / 2

regs_interes = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        )

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    


b=sig.medfilt((sig.medfilt(ecg_one_lead,201)),601)

ecg_filt=ecg_one_lead-b


#FILTRO IIR
    
#FILTRO PASA ALTO
fn=fs/2
attp=0.5
atts=40.0
wppa=3.0/fn
wspa=0.5/fn
filtro1_iirpa=sig.iirdesign(wppa,wspa,gpass=attp,gstop=atts,analog=False,ftype='butter',output='sos')
w , h1_buttpa=sig.sosfreqz(filtro1_iirpa)
ecg_filt1pa=sig.sosfiltfilt(filtro1_iirpa,ecg_one_lead)
wpa = w / np.pi * fn

#FILTRO PASA BAJO
wppb=35/fn
wspb=50/fn
filtro1_iirpb=sig.iirdesign(wppb,wspb,gpass=attp,gstop=atts,analog=False,ftype='butter',output='sos')
w , h1_buttpb=sig.sosfreqz(filtro1_iirpb)
ecg_filt1total=sig.sosfiltfilt(filtro1_iirpb,ecg_filt1pa)
wpb = w / np.pi * fn

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', lw=2)
    plt.plot(zoom_region, ecg_filt[zoom_region], label='filtro  de la mediana')
  

    plt.title('zona de analisis de la mediana')
    plt.ylabel('adimencional')
    plt.xlabel('muestras')
    plt.grid()
    axes_hdl = plt.gca()
    axes_hdl.legend()
    
plt.grid()
axes_hdl = plt.gca()
axes_hdl.legend()

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', lw=2)
    plt.plot(zoom_region, ecg_filt[zoom_region], label='filtro  de mediana')
    plt.plot(zoom_region, ecg_filt1total[zoom_region], label='filtro  IIR')

    plt.title('comparacion filtro mediana con el IIR')
    plt.ylabel('adimencional')
    plt.xlabel('muestras')
    plt.grid()
    axes_hdl = plt.gca()
    axes_hdl.legend()
    
plt.grid()
axes_hdl = plt.gca()
axes_hdl.legend()

plt.show()





```
